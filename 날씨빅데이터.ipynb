```python
import pandas as pd

df = pd.read_csv("cat119_train_train.csv")

df.columns = [
    'index',
    'tm',
    'address_city',
    'address_gu',
    'sub_address',
    'cat',
    'sub_cat',
    'stn',
    'call_count'
]

df.to_csv("cat119_train_cleaned.csv", index=False, encoding='utf-8')
```

```python
import pandas as pd

# 1. 데이터 로드
try:
    df_call = pd.read_csv('call119_train_cleaned.csv')
    df_cat = pd.read_csv('cat119_train_cleaned.csv')
    print("CSV 파일 로드 완료.")
except FileNotFoundError as e:
    print(f"파일을 찾을 수 없습니다: {e}")
    print("스크립트와 같은 경로에 CSV 파일이 있는지 확인해주세요.")
    exit() # 파일이 없으면 스크립트 종료

# 2. 'cat119' 데이터 전처리: 'cat'과 'sub_cat' 조합 컬럼 생성 및 피벗
# 병합 키 컬럼들을 리스트로 정의
merge_keys = [
    'tm',
    'address_city',
    'address_gu',
    'sub_address',
    'stn'
]

# 'cat'과 'sub_cat'을 조합하여 새로운 'combined_cat' 컬럼 생성
# 예: '구급_교통사고', '화재_건물화재'
df_cat['combined_cat'] = df_cat['cat'] + '_' + df_cat['sub_cat']

# 'combined_cat'을 기준으로 피벗하여 각 유형별 신고 건수 컬럼 생성
df_cat_pivot = df_cat.pivot_table(
    index=merge_keys,
    columns='combined_cat', # 'combined_cat' 컬럼의 고유 값들이 새로운 컬럼명이 됨
    values='call_count',
    aggfunc='sum'
).reset_index()

# 피벗 테이블로 인해 생긴 columns.name 제거
df_cat_pivot.columns.name = None

# 새로 생성된 'combined_cat'별 신고 건수 컬럼들의 이름 변경
# 예: '구급_교통사고' -> 'call_count_구급_교통사고'
new_combined_cat_cols = {col: f'call_count_{col}' for col in df_cat_pivot.columns if col not in merge_keys}
df_cat_pivot = df_cat_pivot.rename(columns=new_combined_cat_cols)

print("\n'cat119' 데이터 (combined_cat) 피벗 완료. 생성된 신고 유형별 컬럼 예시:")
print(df_cat_pivot.head())


# 3. 두 데이터프레임 병합
df_merged = pd.merge(df_call, df_cat_pivot, on=merge_keys, how='left')

# 4. 병합 후 NaN 값 처리
combined_cat_call_count_cols = [col for col in df_merged.columns if col.startswith('call_count_') and col != 'call_count']
df_merged[combined_cat_call_count_cols] = df_merged[combined_cat_call_count_cols].fillna(0)


print("\n두 데이터프레임 병합 및 결측치(NaN) 처리 완료.")
print("최종 병합된 데이터의 상위 5행:")
print(df_merged.head())

print("\n최종 데이터프레임의 컬럼 목록:")
print(df_merged.columns.tolist())

print(f"\n최종 데이터프레임의 형태: {df_merged.shape}")

# # 결과를 CSV 파일로 저장 (선택 사항)
# df_merged.to_csv('merged_119_weather_subcat_data.csv', index=False)
# print("\n병합된 데이터를 'merged_119_weather_subcat_data.csv' 파일로 저장했습니다.")
```

```python
import pandas as pd

# 1. 데이터 로드
try:
    df_call = pd.read_csv('call119_train_cleaned.csv')
    df_cat = pd.read_csv('cat119_train_cleaned.csv')
    print("CSV 파일 로드 완료.")
except FileNotFoundError as e:
    print(f"파일을 찾을 수 없습니다: {e}")
    print("스크립트와 같은 경로에 CSV 파일이 있는지 확인해주세요.")
    exit() # 파일이 없으면 스크립트 종료

# 2. 'cat119' 데이터 전처리 및 피벗 (오직 'cat' 컬럼만 사용)
# 병합 키 컬럼들을 리스트로 정의
merge_keys = [
    'tm',
    'address_city',
    'address_gu',
    'sub_address',
    'stn'
]

# 'cat' (신고 유형)을 기준으로 피벗하여 각 유형별 신고 건수 컬럼 생성
# aggfunc='sum'을 사용하여 동일한 기준(tm, 주소, stn)의 'cat'별 'call_count'를 합산
# 이 부분은 첫 번째 답변에서 제공했던 코드와 동일합니다.
df_cat_pivot = df_cat.pivot_table(
    index=merge_keys,
    columns='cat', # 오직 'cat' 컬럼만 사용
    values='call_count',
    aggfunc='sum'
).reset_index()

# 피벗 테이블로 인해 생긴 columns.name 제거 (깔끔한 컬럼명을 위함)
df_cat_pivot.columns.name = None

# 새로 생성된 'cat'별 신고 건수 컬럼들의 이름 변경 (예: '화재' -> 'call_count_화재')
new_cat_cols = {col: f'call_count_{col}' for col in df_cat_pivot.columns if col not in merge_keys}
df_cat_pivot = df_cat_pivot.rename(columns=new_cat_cols)

print("\n'cat119' 데이터 피벗 완료. 생성된 신고 유형별 컬럼 예시:")
print(df_cat_pivot.head())


# 3. 두 데이터프레임 병합
df_merged = pd.merge(df_call, df_cat_pivot, on=merge_keys, how='left')

# 4. 병합 후 NaN 값 처리
# 'call_count_'로 시작하는 새로 추가된 신고 유형별 컬럼들의 NaN 값을 0으로 채움
cat_call_count_cols = [col for col in df_merged.columns if col.startswith('call_count_') and col != 'call_count']
df_merged[cat_call_count_cols] = df_merged[cat_call_count_cols].fillna(0)


print("\n두 데이터프레임 병합 및 결측치(NaN) 처리 완료.")
print("최종 병합된 데이터의 상위 5행:")
print(df_merged.head())

print("\n최종 데이터프레임의 컬럼 목록:")
print(df_merged.columns.tolist())

print(f"\n최종 데이터프레임의 형태: {df_merged.shape}")

# # 결과를 CSV 파일로 저장 (선택 사항)
# df_merged.to_csv('merged_119_weather_cat_only_data.csv', index=False)
# print("\n병합된 데이터를 'merged_119_weather_cat_only_data.csv' 파일로 저장했습니다.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.font_manager as fm # 폰트 설정을 위한 라이브러리

# -------------------------------------------------------------------------
# df_merged_limited_cat 데이터프레임이 이미 생성되어 있다고 가정합니다.
# 실제 사용 시에는 이전에 제공된 병합 코드를 먼저 실행하여 해당 DataFrame을 생성해야 합니다.
# -------------------------------------------------------------------------


# --- 한글 폰트 설정 ---
font_path = None
for font in fm.fontManager.ttflist:
    if 'Malgun Gothic' in font.name: # Windows
        font_path = font.fname
        break
    elif 'AppleGothic' in font.name: # macOS
        font_path = font.fname
        break

if font_path:
    plt.rc('font', family=fm.FontProperties(fname=font_path).get_name())
    print(f"한글 폰트 '{fm.FontProperties(fname=font_path).get_name()}' 설정 완료.")
else:
    print("경고: 적절한 한글 폰트를 찾을 수 없습니다. 폰트 깨짐 현상이 지속될 수 있습니다.")
    print("시스템에 'Malgun Gothic' 또는 'AppleGothic' 폰트가 설치되어 있는지 확인해주세요.")
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

print("\n새로 생성된 df_merged_limited_cat의 컬럼 목록:")
print(df_merged.columns.tolist())

# 시각화할 신고 유형별 컬럼 리스트
target_call_types = [
    'call_count_구급',
    'call_count_구조',
    'call_count_기타',
    'call_count_화재'
]

# 한글 신고 유형 이름을 영어로 매핑 (플롯 제목 및 레이블용)
korean_to_english_map = {
    '구급': 'EMS',
    '구조': 'Rescue',
    '기타': 'Other',
    '화재': 'Fire'
}

# 각 신고 유형별로 최고온도와의 분포 시각화
plt.figure(figsize=(18, 12)) # 전체 그림 크기 조정

for i, col in enumerate(target_call_types):
    if col in df_merged.columns:
        # 컬럼 이름에서 한글 유형명 추출
        korean_type = col.replace('call_count_', '')
        # 영어 이름 매핑
        english_type = korean_to_english_map.get(korean_type, korean_type) # 매핑 없으면 한글 그대로 사용

        plt.subplot(2, 2, i + 1) # 2행 2열 그리드에 플롯
        sns.regplot(x='ta_max', y=col, data=df_merged, scatter_kws={'alpha':0.6})
        plt.title(f'Max Temperature vs. {english_type} Call Count')
        plt.xlabel('Max Temperature (°C)')
        plt.ylabel(f'{english_type} Call Count')
        plt.grid(True, linestyle='--', alpha=0.7)
    else:
        print(f"경고: '{col}' 컬럼이 df_merged_limited_cat에 없습니다. 병합 코드를 확인하세요.")

plt.tight_layout() # 플롯 간 간격 자동 조절
plt.show()

# 추가: 각 유형별 최고기온과의 상관계수 출력
print("\n--- 최고기온과 각 신고 유형별 건수의 상관계수 ---")
for col in target_call_types:
    if col in df_merged.columns:
        correlation = df_merged['ta_max'].corr(df_merged[col])
        korean_type = col.replace('call_count_', '')
        english_type = korean_to_english_map.get(korean_type, korean_type)
        print(f"최고기온(Max Temp)과 {english_type} Call Count 상관계수: {correlation:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# 학습 데이터 로드
train = pd.read_csv("call119_train_cleaned.csv")  
X_train = train[['ta_max', 'ta_min', 'ta_max_min', 'hm_min', 'hm_max',
                 'ws_max', 'ws_ins_max', 'rn_day']]
y_train = train['call_count']

# 모델 훈련
model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)

# 테스트 데이터 불러오기
test = pd.read_csv("test_call119.csv", encoding='cp949')
X_test = test[X_train.columns]  # 동일한 피처 사용

# 예측 및 정수화
test['call_count'] = model.predict(X_test).round().astype(int)

# 제출용 저장
test.to_csv("123456.csv", index=False, encoding='cp949')  # 참가번호로 저장
```

```python
from sklearn.metrics import mean_squared_error

# 실제 정답 백업
y_true = test['call_count']  # 기존 값 (예: 0 또는 채워져 있는 정답)
# 예측값
y_pred = model.predict(X_test)

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"RMSE: {rmse:.4f}")
```
